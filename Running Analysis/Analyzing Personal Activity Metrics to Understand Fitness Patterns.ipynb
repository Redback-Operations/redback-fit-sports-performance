```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Fixed_cleaned_activities.csv')

# Convert time columns to timedelta
df['Duration (seconds)'] = pd.to_timedelta(df['Duration (h:m:s)']).dt.total_seconds()
df['Moving Duration (seconds)'] = pd.to_timedelta(df['Moving Duration (h:m:s)']).dt.total_seconds()

# Convert Begin Timestamp to datetime
df['Begin Timestamp'] = pd.to_datetime(df['Begin Timestamp'])

# Basic Overview
print("=== Dataset Overview ===")
print(df.describe())
```

# Heart rate trends and how they change over time

```python
# 1. Plot Average Heart Rate Over Time
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Begin Timestamp', y='Average Heart Rate (bpm)')
plt.title('Average Heart Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Avg Heart Rate (bpm)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Elevations impact on calories burned

```python
# 2. Elevation Gain vs Calories Burned
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Elevation Gain (Raw)', y='Calories', hue='Activity Type')
plt.title('Elevation Gain vs Calories Burned')
plt.xlabel('Elevation Gain (m)')
plt.ylabel('Calories')
plt.tight_layout()
plt.show()
```

```python
# Line Plot 4: Average Speed 
sns.lineplot(x='Begin Timestamp', y='Average Speed', hue='Activity Type', data=df)
plt.title('Average Speed Over Time')
plt.xlabel('Date')
plt.ylabel('Average Speed (m/s)')
plt.tight_layout()
plt.show()
```

# Elevation Gain and Calories Burned Over Time

```python
# Line Plot 3: Elevation Gain vs Calories Over Time
sns.lineplot(x='Begin Timestamp', y='Elevation Gain (Raw)', data=df, label='Elevation Gain')
sns.lineplot(x='Begin Timestamp', y='Calories', data=df, label='Calories Burned')
plt.title('Elevation Gain and Calories Burned Over Time')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.tight_layout()
plt.show()
```
