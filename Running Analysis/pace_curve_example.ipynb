## **Pace Curve Analyser Example**

The purpose of this notebook is to provide an example of how to implement the Pace Curve Analyser.

```python
import pandas as pd
from pace_curve import PaceCurveAnalyzer

# Example DataFrame (replace this with your actual data)
# Ensure your data contained the following attributes
data = {
    'Activity Date': ['2023-05-01', '2023-05-02', '2023-05-03'],
    'Activity Type': ['Run', 'Run', 'Run'],
    'Distance': [5, 10, 15],  # in kilometers
    'Elapsed Time': [1500, 3600, 5400]  # in seconds
}

# Convert the data to a DataFrame
df = pd.DataFrame(data)

# Instantiate the PaceCurveAnalyser with the data
pca = PaceCurveAnalyzer(data_source=df)

# Create a pace curve for 'Run' activities within the last 90 days ending on '2023-05-30'.
# Change the input parameters to fit your requirements.
power_curve = pca.create_pace_curve('Run', '2023-05-30', 90)

# Plot the pace curve
pca.plot_pace_curve(power_curve)
```

```python
# Add the pace curve column to the DataFrame
df_with_pace_curve = pca.add_pace_curve_column(activity_type='Run', num_days=90)

# Expand the pace curve into individual distance columns
expanded_df = pca.expand_pace_curve_to_columns()
expanded_df
```
