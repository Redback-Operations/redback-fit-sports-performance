```python
# ============================================================
# ReflexionPro — Muscle Activation Tracking (EMG end‑to‑end)
# Load → Preprocess → Rep segmentation → Per‑rep metrics
# → Window dataset → Model (time split / GroupKFold) → Save
# ============================================================

# -------- 0) Imports & config
import os, re, glob, json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from io import StringIO
from scipy.signal import butter, filtfilt, savgol_filter, find_peaks
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GroupKFold
from sklearn.metrics import mean_absolute_error, r2_score

plt.rcParams['figure.figsize'] = (10, 6)

# >>> SET ME: folder containing your *.txt EMG files (can be the top-level; search is recursive)
DATA_DIR = "/Users/chirugowda/Desktop/phythom_projects/SEMG_DB1 2"
# optional: choose which file to use from the discovered list (0 = longest)
SELECT_INDEX = 0

# EMG signal config
FS = 1000            # Hz (dataset spec)
BP_LOW, BP_HIGH = 20, 450
ENV_LP = 6           # Hz (envelope smoothing)
ROBUST_Q = 95        # percentile for robust "max" normalization

# -------- 1) Robust reader for N_TXT
_num_re = re.compile(r'[-+]?\d*\.?\d+(?:[eE][-+]?\d+)?')

def _extract_numeric_tokens(line: str):
    # convert decimal commas to dots only inside numbers
    line = re.sub(r'(?<=\d),(?=\d)', '.', line)
    return _num_re.findall(line)

def read_emg_txt(path: str) -> pd.DataFrame:
    """
    Robustly read an EMG txt file:
      - skips any text header
      - auto-detects the first numeric row
      - accepts tabs/spaces/mixed delimiters
      - returns columns: RF, BF, VM, ST, FX
    """
    with open(path, 'r', errors='ignore') as f:
        lines = f.readlines()

    data_start = None
    for i, line in enumerate(lines):
        if len(_extract_numeric_tokens(line)) >= 5:
            data_start = i; break
    if data_start is None:
        raise ValueError(f"No numeric block found: {path}")

    numeric_lines = []
    for line in lines[data_start:]:
        toks = _extract_numeric_tokens(line)
        if len(toks) >= 5:
            numeric_lines.append(' '.join(toks[:5]))

    if not numeric_lines:
        raise ValueError(f"Numeric start found but no rows parsed: {path}")

    arr = np.loadtxt(StringIO('\n'.join(numeric_lines)))
    if arr.ndim == 1:
        arr = arr.reshape(-1, 5)

    df = pd.DataFrame(arr, columns=['RF','BF','VM','ST','FX'])
    return df

# -------- 2) Safe filters + envelope + %activation
def _safe_filtfilt(b, a, x):
    x = np.asarray(x, float)
    padlen_needed = 3 * (max(len(a), len(b)) - 1)
    if x.size <= padlen_needed + 1:
        return filtfilt(b, a, x, method="gust")  # handles short signals
    return filtfilt(b, a, x)

def bandpass(sig, fs=FS, low=BP_LOW, high=BP_HIGH, order=4):
    b, a = butter(order, [low/(fs/2), high/(fs/2)], btype='band')
    return _safe_filtfilt(b, a, sig)

def lowpass(sig, fs=FS, cutoff=ENV_LP, order=4):
    b, a = butter(order, cutoff/(fs/2), btype='low')
    return _safe_filtfilt(b, a, sig)

def emg_envelope(raw, fs=FS):
    raw = np.asarray(raw, float)
    bp  = bandpass(raw, fs)
    rect = np.abs(bp)
    env = lowpass(rect, fs)
    return env

def to_percent_activation(env, q=ROBUST_Q):
    ref = np.percentile(env, q)
    return (env / (ref + 1e-9)) * 100.0

def preprocess_file(path):
    df = read_emg_txt(path)
    env = {ch: emg_envelope(df[ch].values, FS) for ch in ['RF','BF','VM','ST']}
    pct = {ch: to_percent_activation(env[ch]) for ch in env}
    pct_df = pd.DataFrame(pct)
    pct_df['FX'] = df['FX'].values
    return df, pct_df

# -------- 3) Robust file discovery (recursive) + pick longest file
def list_txt_files_recursive(data_dir):
    return sorted(glob.glob(os.path.join(data_dir, "**", "*.txt"), recursive=True))

def try_parse_length(path):
    try:
        return len(read_emg_txt(path))
    except Exception:
        return 0

all_txt = list_txt_files_recursive(DATA_DIR)
print(f"Found .txt files (recursive): {len(all_txt)}")

parsable = [(p, try_parse_length(p)) for p in all_txt]
parsable = [(p, n) for (p, n) in parsable if n > 0]

if not parsable:
    raise RuntimeError("No parsable .txt files under DATA_DIR. Check path and extraction.")

parsable.sort(key=lambda x: x[1], reverse=True)
print("\nParsable files (top 10 by length):")
for p, n in parsable[:10]:
    print(f" - {os.path.basename(p)}  rows={n}")

SELECT_INDEX = max(0, min(SELECT_INDEX, len(parsable)-1))
file_path = parsable[SELECT_INDEX][0]
print(f"\n Using file: {os.path.basename(file_path)}")

raw_df, pct_df = preprocess_file(file_path)
print("Rows:", len(raw_df))

# -------- 4) Rep segmentation from FX (flexion cycles)
def segment_reps_from_fx(fx, fs=FS, min_rep_dur_s=0.5, peak_height_q=60, pad_s=0.10):
    fx = pd.Series(fx, dtype=float).interpolate().bfill().ffill().values
    win = 51 if len(fx) > 51 else max(5, (len(fx)//2)*2-1)
    fx_s = savgol_filter(fx, window_length=win, polyorder=2, mode="interp")
    fx_vel = np.gradient(fx_s)

    distance = int(fs * min_rep_dur_s)
    height = np.percentile(np.abs(fx_vel), peak_height_q)
    peaks, _ = find_peaks(np.abs(fx_vel), distance=distance, height=height)

    pad = int(pad_s * fs)
    bounds = []
    for i in range(0, len(peaks)-1, 2):
        s = max(0, peaks[i] - pad)
        e = min(len(fx), peaks[i+1] + pad)
        if e - s >= distance:
            bounds.append((s, e))
    return bounds

# try defaults; if none, relax thresholds
rep_bounds = segment_reps_from_fx(raw_df['FX'].values, fs=FS, min_rep_dur_s=0.5, peak_height_q=60, pad_s=0.10)
if len(rep_bounds) == 0:
    rep_bounds = segment_reps_from_fx(raw_df['FX'].values, fs=FS, min_rep_dur_s=0.35, peak_height_q=40, pad_s=0.08)

print("Detected reps:", len(rep_bounds))
if rep_bounds[:3]: print("First 3 reps:", rep_bounds[:3])

# -------- 5) Per‑rep metrics
def rep_metrics(pct_df, s, e, muscles=('RF','BF','VM','ST')):
    out = {'t_start_s': s/FS, 't_end_s': e/FS, 'duration_s': (e-s)/FS}
    for m in muscles:
        seg = pct_df[m].values[s:e]
        out[f'{m}_mean']      = float(np.nanmean(seg))
        out[f'{m}_peak']      = float(np.nanmax(seg))
        out[f'{m}_iEMG']      = float(np.nansum(seg) / FS)    # %·s
        out[f'{m}_t_above40'] = float((seg > 40).sum() / FS)  # sec
    return out

rep_rows = [rep_metrics(pct_df, s, e) for (s, e) in rep_bounds]
rep_table = pd.DataFrame(rep_rows)
print("rep_table shape:", rep_table.shape)
display(rep_table.head(10) if not rep_table.empty else rep_table)

# -------- 6) Window‑level dataset (predict VM from RF/BF/ST)
def sliding_windows(length, win, hop):
    for s in range(0, length - win + 1, hop):
        yield s, s + win

def feats_1d(arr):
    arr = np.asarray(arr, float)
    return {
        "mean": float(np.nanmean(arr)),
        "rms":  float(np.sqrt(np.nanmean(arr**2))),
        "var":  float(np.nanvar(arr)),
        "wl":   float(np.nansum(np.abs(np.diff(arr))))
    }

TARGET_MUSCLE = "VM"
PRED_CH = ['RF','BF','ST']

def build_window_data(pct_df, target=TARGET_MUSCLE, preds=PRED_CH, win_ms=200, hop_ms=50):
    win = int(FS * win_ms / 1000.0)
    hop = int(FS * hop_ms / 1000.0)
    rows, y, idxs = [], [], []
    L = len(pct_df)
    for s, e in sliding_windows(L, win, hop):
        feat = {}; ok = True
        for ch in preds:
            w = pct_df[ch].values[s:e]
            if not np.isfinite(w).all() or len(w) < 2:
                ok = False; break
            feat.update({f"{ch}_{k}": v for k, v in feats_1d(w).items()})
        if not ok: continue
        rows.append(feat)
        y.append(float(np.nanmean(pct_df[target].values[s:e])))
        idxs.append((s, e))
    X = pd.DataFrame(rows); y = np.array(y, float)
    return X, y, idxs

WIN_MS, HOP_MS = 200, 50
X, y, idxs = build_window_data(pct_df, TARGET_MUSCLE, PRED_CH, win_ms=WIN_MS, hop_ms=HOP_MS)

# auto‑shrink windows for very short files so you still get data
if len(X) == 0:
    WIN_MS, HOP_MS = 100, 25
    X, y, idxs = build_window_data(pct_df, TARGET_MUSCLE, PRED_CH, win_ms=WIN_MS, hop_ms=HOP_MS)

print(f"Windows: {len(X)} | Features: {X.shape[1] if len(X) else 0} | (win={WIN_MS}ms, hop={HOP_MS}ms)")
display(X.head() if not X.empty else X)

# -------- 7) Train & evaluate (time split). If many files exist, also GroupKFold.
def time_split_eval(X, y, ratio=0.7):
    if len(X) < 10:
        raise ValueError("Not enough windows to split. Collect a longer file or shrink WIN_MS/HOP_MS.")
    split = int(ratio * len(X))
    X_tr, X_te = X.iloc[:split], X.iloc[split:]
    y_tr, y_te = y[:split], y[split:]

    rf = RandomForestRegressor(
        n_estimators=600, max_depth=None,
        min_samples_split=4, min_samples_leaf=2,
        random_state=42, n_jobs=-1
    )
    rf.fit(X_tr, y_tr)
    y_hat = rf.predict(X_te)
    mae = mean_absolute_error(y_te, y_hat)
    r2  = r2_score(y_te, y_hat)
    return rf, (mae, r2), (y_te, y_hat)

rf, (mae_ts, r2_ts), (y_te, y_hat) = time_split_eval(X, y, ratio=0.7)
print(f"[Time-split] Test MAE: {mae_ts:.2f} | R²: {r2_ts:.3f}")

# GroupKFold across files if you have several trials
other_files = [p for p, _ in parsable]  # all parsable
if len(other_files) > 1:
    X_all, y_all, groups = [], [], []
    for p in other_files:
        _, pct_f = preprocess_file(p)
        Xf, yf, _ = build_window_data(pct_f, TARGET_MUSCLE, PRED_CH, win_ms=WIN_MS, hop_ms=HOP_MS)
        if len(Xf) == 0:
            continue
        X_all.append(Xf); y_all.append(yf); groups += [os.path.basename(p)]*len(Xf)
    if X_all:
        Xg = pd.concat(X_all, ignore_index=True)
        yg = np.concatenate(y_all)
        groups = np.array(groups)

        n_groups = len(np.unique(groups))
        n_splits = 5 if n_groups >= 5 else max(2, n_groups)
        gkf = GroupKFold(n_splits=n_splits)

        fold_rows = []
        for fold, (tr, te) in enumerate(gkf.split(Xg, yg, groups=groups), 1):
            model = RandomForestRegressor(
                n_estimators=600, max_depth=None,
                min_samples_split=4, min_samples_leaf=2,
                random_state=42, n_jobs=-1
            )
            model.fit(Xg.iloc[tr], yg[tr])
            yh = model.predict(Xg.iloc[te])
            fold_rows.append({"fold":fold,
                              "MAE": mean_absolute_error(yg[te], yh),
                              "R2":  r2_score(yg[te], yh)})
        cv_df = pd.DataFrame(fold_rows)
        print("\n[GroupKFold by file]")
        print(cv_df)
        print("Averages:\n", cv_df.mean(numeric_only=True))

# -------- 8) Plots (time split)
plt.figure()
plt.scatter(y_te, y_hat, s=10)
plt.xlabel(f"True {TARGET_MUSCLE} %activation")
plt.ylabel(f"Pred {TARGET_MUSCLE} %activation")
plt.title("Window-level prediction (time-based split)")
plt.show()

imp = pd.Series(rf.feature_importances_, index=X.columns).sort_values(ascending=False)
plt.figure()
imp.head(12).plot(kind="bar")
plt.title("Top feature importances"); plt.ylabel("Importance")
plt.show()

# -------- 9) Save outputs
os.makedirs("outputs", exist_ok=True)
rep_csv   = os.path.join("outputs", "rep_metrics.csv")
win_csv   = os.path.join("outputs", "window_dataset_preview.csv")
model_pkl = os.path.join("outputs", f"rf_model_target_{TARGET_MUSCLE}.joblib")

rep_table.to_csv(rep_csv, index=False)
X.head(200).to_csv(win_csv, index=False)

try:
    import joblib
    joblib.dump(rf, model_pkl)
    print("Saved model:", model_pkl)
except Exception as e:
    print("Skipping model save:", e)

print("Saved:", rep_csv)
print("Saved:", win_csv)
```
