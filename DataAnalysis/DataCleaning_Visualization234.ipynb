```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load CSV file
df = pd.read_csv("activities_cleaned 1.csv")

# Convert the timestamp columns to datetime format
df['Begin Timestamp'] = pd.to_datetime(df['Begin Timestamp'], errors='coerce')
df['End Timestamp'] = pd.to_datetime(df['End Timestamp'], errors='coerce')

# Calculate workout duration in minutes
df['Workout Duration (min)'] = (df['End Timestamp'] - df['Begin Timestamp']).dt.total_seconds() / 60
```

```python
# Drop rows with missing timestamps
df_cleaned = df.dropna(subset=['Begin Timestamp', 'End Timestamp'])

# Copy for visualization
df_viz = df_cleaned.copy()
```

```python
# Plot workout duration by weather condition
plt.figure(figsize=(10, 6))
sns.barplot(data=df_viz, x='Condition', y='Workout Duration (min)', ci=None)
plt.xticks(rotation=45)
plt.title('Average Workout Duration by Weather Condition')
plt.ylabel('Duration (minutes)')
plt.xlabel('Weather Condition')
plt.tight_layout()
plt.show()
```

This week, I focused on visualising key aspects of the cleaned workout dataset. I created a bar chart to show the relationship between workout duration and weather condition, and a scatter plot to examine the correlation between distance and average speed.
