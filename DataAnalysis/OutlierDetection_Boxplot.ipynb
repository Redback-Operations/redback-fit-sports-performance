```python
# Import libraries
import pandas as pd

# Load CSV file
df = pd.read_csv("activities_cleaned 1.csv")

# Convert timestamp columns to datetime
df['Begin Timestamp'] = pd.to_datetime(df['Begin Timestamp'], errors='coerce')
df['End Timestamp'] = pd.to_datetime(df['End Timestamp'], errors='coerce')

# Calculate duration in minutes
df['Workout Duration (min)'] = (df['End Timestamp'] - df['Begin Timestamp']).dt.total_seconds() / 60

# Drop rows with missing timestamps
df_cleaned = df.dropna(subset=['Begin Timestamp', 'End Timestamp'])
```

```python
# Import required libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Create boxplot of workout duration
plt.figure(figsize=(8, 5))  # Set the figure size
sns.boxplot(data=df_cleaned, y='Workout Duration (min)')  # Create the boxplot
plt.title("Boxplot of Workout Duration (min)")  # Add plot title
plt.ylabel("Duration (minutes)")  # Set y-axis label
plt.grid(True)  # Show grid
plt.show()  # Display the plot
```

```python
# Workout Duration 
df_filtered = df_cleaned[df_cleaned['Workout Duration (min)'] <= 1000]

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.boxplot(data=df_filtered, y='Workout Duration (min)')
plt.title("Boxplot of Workout Duration (min)")
plt.ylabel("Duration (minutes)")
plt.grid(True)
plt.show()
```

In Week 5, we used boxplots to analyze workout durations and identified extreme outliers. After filtering out durations over 1000 minutes, the refined visualization showed a more realistic distribution
