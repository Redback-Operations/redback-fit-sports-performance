```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("activities_cleaned 1.csv")  # Ensure this file is in the same directory

# Convert timestamp columns to datetime format
df['Begin Timestamp'] = pd.to_datetime(df['Begin Timestamp'], errors='coerce')
df['End Timestamp'] = pd.to_datetime(df['End Timestamp'], errors='coerce')

# Calculate workout duration in minutes
df['Workout Duration (min)'] = (df['End Timestamp'] - df['Begin Timestamp']).dt.total_seconds() / 60

# Show basic info
df.info()
df.head()
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Convert timestamp columns to datetime format (already checked in your df.info())
df['Begin Timestamp'] = pd.to_datetime(df['Begin Timestamp'], errors='coerce')
df['End Timestamp'] = pd.to_datetime(df['End Timestamp'], errors='coerce')

# Calculate workout duration in minutes
df['Workout Duration (min)'] = (df['End Timestamp'] - df['Begin Timestamp']).dt.total_seconds() / 60

# Plot average heart rate under different weather conditions
plt.figure(figsize=(8, 6))
df.boxplot(column='Average Heart Rate (bpm)', by='Condition')  # Column name must be exactly 'Condition'
plt.title('Heart Rate vs. Weather Condition')
plt.suptitle('')  # Remove the default title
plt.xlabel("Weather Condition")
plt.ylabel("Average Heart Rate (bpm)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

In this Week, a boxplot was successfully created to visualize the variation of average heart rate under different weather conditions using the Condition and Average Heart Rate (bpm) columns. The timestamps were correctly converted to datetime format, and workout duration was computed in minutes.
