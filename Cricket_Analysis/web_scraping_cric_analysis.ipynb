#### Listing the urls correspoding to each anchor tag in the match_summary table

```python
import requests
from bs4 import BeautifulSoup

# Replace 'url' with the actual URL of the webpage containing the HTML table
url = 'https://www.espncricinfo.com/records/tournament/team-match-results/icc-cricket-world-cup-2023-24-15338'

# Send an HTTP request to the webpage and get the HTML content
response = requests.get(url)
html_content = response.text

# Create a BeautifulSoup object to parse the HTML
soup = BeautifulSoup(html_content, 'html.parser')

# Locate the HTML table using appropriate tags and attributes
# For example, if the table has the id 'my-table', you can use: table = soup.find('table', {'id': 'my-table'})
table = soup.find('table')

# Initialize a list to store the extracted links
links = []

# Iterate through each row in the table
for row in table.find_all('tr'):
    # Find all the anchor tags (links) in each row
    for link in row.find_all('a', href=True):
        # Get the value of the 'href' attribute (link)
        href = link['href']
        # Append the link to the 'links' list
        links.append(href)

# Print the collected links
for link in links:
    print(link)
```

#### Filtering the urls correspoding to the ground information alone

```python
links1 = links[::2]
```

#### Finally filtering the urls of all the 48 matches in the world cup

```python
# Use list comprehension to create a new list without elements from list2
result_list = []
for item in links:
    if item not in links1:
               result_list.append(item)

print(result_list)
```
